```python
from google.colab import files
uploaded = files.upload()
```

```python
import zipfile
import io
z = zipfile.ZipFile(io.BytesIO(uploaded['spam_ham_dataset.csv.zip']))
df = pd.read_csv(z.open('spam_ham_dataset.csv'))
df.head()
```

```python
df_200 = df.head(200)
df_200['label_num'] = df_200['label'].map({'spam':1, 'ham':0})
X_text = df_200['text']
y = df_200['label_num']
```

```python
X_train_text = X_text[:150]
X_test_text = X_text[150:]
y_train = y[:150]
y_test = y[150:]
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X_train = vectorizer.fit_transform(X_train_text)
X_test = vectorizer.transform(X_test_text)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_class = (y_pred > 0.5).astype(int)
accuracy = accuracy_score(y_test, y_pred_class)
accuracy
```

```python
results = pd.DataFrame({
    'Message': X_test_text.values,
    'Actual': y_test.values,
    'Predicted': y_pred_class
})
results.head(10)
```
